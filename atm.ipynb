```python
print("Welcome to Simple ATM Simulator!")
print("\n c: Abid Hasan")

balance = 10000
user_pin = '2010'
enter_pin = input("Enter your PIN: ")

if enter_pin != user_pin:
    print("Invalid PIN!")
    atm_on = False
else:
    atm_on = True

while atm_on:
    print("\nMain Menu: ")
    print("1. Check Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Exit")

    choice = input("Enter your choice: ")

    # Make sure choice is an integer
    if choice.isdigit():
        choice = int(choice)
    else:
        print("Invalid input! Please enter a number.")
        continue

    if choice == 1:
        print("\n Your current balance is: $" + str(balance))
    elif choice == 2:
        amount = float(input("\n Deposit Amount: $"))
        balance += amount
        print(" \n $" + str(amount) + " Deposit Successful!")
    elif choice == 3:
        amount = float(input("Withdraw Amount: $"))
        if amount > balance:
            print("Insufficient Balance!")
        else:
            balance -= amount
            print("\n $" + str(amount) + " Withdraw Successful!")
    elif choice == 4:
        print("Thank you for using our ATM. Goodbye!")
        atm_on = False
    else:
        print("Invalid Choice! Please try again.")
```
